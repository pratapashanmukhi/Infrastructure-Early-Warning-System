import pandas as pd

bridge = pd.read_csv("bridge.csv")
water = pd.read_csv("water.csv")

print(bridge.head())
print(water.head())

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
bridge["Material_Type"] = le.fit_transform(bridge["Material_Type"])
bridge["Maintenance_Level"] = le.fit_transform(bridge["Maintenance_Level"])

bridge = bridge.drop(columns=["infrastructure_type"])
water = water.drop(columns=["Sensor_ID", "infrastructure_type"])


X_bridge = bridge.drop(columns=["failure"])
y_bridge = bridge["failure"]

X_water = water.drop(columns=["failure"])
y_water = water["failure"]


from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Bridge
Xb_train, Xb_test, yb_train, yb_test = train_test_split(X_bridge, y_bridge, test_size=0.2)
bridge_model = RandomForestClassifier()
bridge_model.fit(Xb_train, yb_train)
print("Bridge accuracy:", bridge_model.score(Xb_test, yb_test))

# Water
Xw_train, Xw_test, yw_train, yw_test = train_test_split(X_water, y_water, test_size=0.2)
water_model = RandomForestClassifier()
water_model.fit(Xw_train, yw_train)
print("Water accuracy:", water_model.score(Xw_test, yw_test))


print(bridge_model.predict([[70, 4000, 1, 0]]))
print(water_model.predict([[3.5, 180, 25, 0]]))
